```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
atribut = iris.data
label = iris.target
```

```python
# bagi dataset menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2)
```

```python
# Coba dengan decision tree
from sklearn import tree

clf_tree = tree.DecisionTreeClassifier().fit(X_train, y_train)
```

```python
# Cek skor
clf_tree.score(X_test, y_test)
```

#Aplikasi PCA

```python
from sklearn.decomposition import PCA
 
# membuat objek PCA dengan 4 principal component
pca = PCA(n_components=4)
```

```python
# mengaplikasikan PCA pada dataset
pca_attributes = pca.fit_transform(X_train)
```

```python
# melihat variance dari setiap atribut
pca.explained_variance_ratio_
```

###Dari hasil di atas, PC pertama memiliki variance yang sangat tinggi, berarti jauh lebih signifikan

```python
# Ambil dua PC saja
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
```

```python
# uji akurasi classifier

# Buat pca untuk x test
X_test_pca = pca.fit_transform(X_test)

clf_tree_pca = tree.DecisionTreeClassifier().fit(X_train_pca, y_train)
clf_tree_pca.score(X_test_pca, y_test)
```

#Kesimpulan:
Walaupun akurasinya lebih rendah, tetapi model yang kedua ini hanya menggunakan dua PC saja. Hal ini akan terasa efeknya ketika menggunakan datasets yang sangat-sangat besar, PCA dapat mengurangi waktu dan memori penggunaan, walaupun terkadang harus mengorbankan akurasi.
