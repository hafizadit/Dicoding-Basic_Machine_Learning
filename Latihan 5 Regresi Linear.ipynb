# Tujuan
Pada latihan kali ini kita akan memprediksi harga rumah berdasarkan jumlah kamar.
# Tahapan
1. Import library yang dibutuhkan.
2. Buat dataset dummy dengan Numpy Array.
3.Buat plot dari model

```python
import numpy as np

# Buat data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# Buat data harga, asumsi dalam dollar
price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])
```

```python
bedrooms
```

```python
# Plot in scatterplot
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(bedrooms, price)
plt.title("Number of Bedrooms - Price")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price (in $)")
plt.show()
```

```python
# Get the linear regression
from sklearn.linear_model import LinearRegression

bedrooms = bedrooms.reshape(-1,1)

# Train the model
linear_regression = LinearRegression()
linear_regression.fit(bedrooms, price)
```

```python
# Visual the best fit line
plt.scatter(bedrooms, price)
plt.plot(bedrooms, linear_regression.predict(bedrooms))

plt.title("Number of Bedrooms - Price")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price (in $)")

plt.show()
```

```python
# Know the R Squared Value
from sklearn.metrics import mean_squared_error, r2_score

y_pred = linear_regression.predict(bedrooms)
y_true = price

print(f"R^2 Score : {r2_score(y_true, y_pred)}")
```
