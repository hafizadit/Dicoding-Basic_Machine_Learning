##Terdapat data kategorikal (Gender) yang harus diubah dengan One-Hot Encoding


```python
data = pd.get_dummies(df)
```

```python
data.head()
```

```python
# Memisahkan atribut dan label

X = data[["Age","EstimatedSalary","Gender_Female","Gender_Male"]]
y = data["Purchased"]

X.head()
```

#Melakukan normalisasi data dengan **Standardization**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

scaled_data = scaler.transform(X) #This is an array

#Convert to DataFrame
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
```

```python
scaled_data.head()
```

#Split the Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

#Make a Model

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Evaluate the score

for i in range(10):
  print(model.score(x_test,y_test))
```
